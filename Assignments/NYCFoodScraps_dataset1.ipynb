```python
from shapely.geometry import Point
import geopandas as gpd

# Ensure no missing values in coordinates
pluto = pluto.dropna(subset=['Longitude', 'Latitude'])

# Create geometry column
pluto['geometry'] = pluto.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)

# Convert to GeoDataFrame
pluto_gdf = gpd.GeoDataFrame(pluto, geometry='geometry', crs='EPSG:4326')
```

```python
# Plot a map with food scrap drop-off sites by Borough

fig, ax = plt.subplots(figsize=(12, 10))

pluto_gdf.plot(ax=ax, column='Borough', legend=True, markersize=25, alpha=0.7, edgecolor='black')

# Title and labels
plt.title("Food Scrap Drop-off Sites by Borough", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
